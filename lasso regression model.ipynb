```python
def correlation(dataset, threshold):   # create function 
    col_corr = set()                   # seat value function 
    corr_matrix = dataset.corr()       # dataset(X_train) coreleation
    for i in range(len(corr_matrix.columns)):    # store the values which are graeater than threshold value(0.85)
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:   
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)     # threshold = 0.85 
```

```python
corr_features
```

```python
## drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

##Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

Box Plots To understand Effect Of Standard Scaler

```python
plt.subplots(figsize=(20, 10))
plt.subplot(1, 2, 1)                   # 1 row 2 column 1st box
sea.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sea.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linreg=LinearRegression()
```

```python
linreg.fit(X_train_scaled,y_train)
```

```python
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_pred,y_test)
score = r2_score(y_pred,y_test)
print("mae =", mae)
print("score =",score)
```

LASSO REGRESSION 

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
#lasso.fit(X_train_scaled,y_train)
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

Ridge Regression model


```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

Elasticnet Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic=ElasticNet()
elastic.fit(X_train_scaled,y_train)
y_pred=elastic.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```
